### 4.1.3 用不用show()?如何显示图形

**在脚本中画图**  
必须使用show() 将会启动一个事件循环、并且找到所有当前可用的图形对象，
然后打开一个或者多个交互式显示图形

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```shell
$ python myplot.py
```

- 需要注意的是一个python会话只能使用一次show 否则将会出现难以预料的异常

**在shell中画图**  
启动了ipython 后 仅需要使用魔法命令 %matplotlib  
此后任意的plt命令都会自动打开一个图形窗口，增加新的命令图形就会更新
对于某些不及时更新的变化可以使用`plt.draw()` 进行强制更新 就不在需要show

**在notebook中画图**  

有两种展现方式  
- %matplotlib notebook 交互式图形
- %matplotlib inline   静态图形

一般以第二种为主，且也仅需运行一次

````python
%matplotlib inline
x=np.linspace(0,10,100)
fig=plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
````

## 4.2 两种画图接口
- 便捷的MATLAB风格
- 功能更强大的面向对象接口

### 4.2.1 MATLAB风格接口

- pyplot

该接口最重要的特性是**有状态的**：它会持续跟踪当前的图形和坐标轴

- plt.gcf() 获取当前图形
- plt.gca() 获取当前坐标轴

````python
plt.figure()
plt.subplot(211)
plt.plot(x,np.sin(x),'-')
plt.subplot(212)
plt.plot(x,np.cos(x),'--')
````

### 4.2.2 面向对象接口

面向对象接口可以适应更加复杂的场景，控制好自己的图形，通过显示figure & axes方法


````python
fig,ax=plt.subplots(2)
ax[0].plot(x,np.sin(x),'-')
ax[1].plot(x,np.cos(x),'-')
````

## 4.3简易线形图


````python
'''
figure 可以看做是一个能够容纳各种坐标轴、图形文字、标签的容器
axes   是一个带有刻度和标签的矩形
'''
plt.style.use('seaborn-whitegrid')
fig,ax=plt.subplots(1)
ax.plot(x,np.sin(x),'-')
````
